## <font style="color:rgb(134,19,348)">Contrast Enhancement</font>

Contrast can be defined as the difference in intensity between the brightest and darkest regions of a an image. Greater the difference, higher is the contrast. With good Contrast in an image you can easily perceive the objects in it. <br> 

**Note:** Max contrast of an image is known as Dynamic Range.

An easy way to improve or increase contrast is to scale the intensity, so what we do is we multiply all the intensity values of image by a constant factor and add that to the image. 

- $Ib$ Output Image
- $I$  Input Image
- $\beta$ is the Constant factor

$$
I_b = I * \beta
$$

```python
image = cv2.imread('media/M3/lowcont.png')

# Set the percent of contrast you want, in this case we are inreasing the contrast by 25%
Contrast_percent = 55

high_contrast = image * (1 + Contrast_percent / 100)

# Remember that numpy rounds of, so we need to clip values above 255 and below 0 and then convert it to uint8
high_contrast = np.clip(high_contrast, 0, 255)
high_contrast = np.uint8(high_contrast)

# Alternatively you can also divide by 255 and convert to float.

plt.figure(figsize=[18,18])
plt.subplot(131);plt.imshow(image[...,::-1]);plt.title("Normal");
plt.subplot(132);plt.imshow(high_contrast[...,::-1]);plt.title("High Contrast");
```

**As you can see the image on the right is more sharper since it has high contrast**
<br>

Now what this is doing do is that its increasing the difference between intensities. For example if previously the difference between certain two pixels <i>(pixel 1 = 50 and pixel 2 = 100)</i> was 50  and then if we increase the contrast by a 100% , then the pixel 1 and 2 become 100 and 200 respectively and the difference increases to 100.

## <font style="color:rgb(134,19,348)">Brightness Enhancement</font>


In very simple terms you can say that brightness is the measure of how light or dark an image is but brightness is properly defined as a measure of light falling on the image. It can be thought of as the a mean of the R, G and B color values in an RGB image. To make an image brighter you must increase the intensity values by some constant and vice versa.

- $I_b$ Output Image
- $I$  Input Image
- $\beta$ is the Constant 

$$
I_b = I + \beta
$$

```python
image = cv2.imread('media/M3/bright.jpg')

# Set some brightness constant value
constant = 100

# Add constant for increasing brightness, we convert to uint32 to prevent numpy rounding
bright_image = np.int32(image) + constant

# Remember that numpy rounds of, so we need to clip values above 255 and below 0 and then convert it to uint8
bright_image = np.clip(bright_image, 0, 255)
bright_image = np.uint8(bright_image)

plt.figure(figsize=[18,18])
plt.subplot(131);plt.imshow(image[...,::-1]);plt.title("Normal");plt.axis('off');
plt.subplot(132);plt.imshow(bright_image[...,::-1]);plt.title("High Brightness");plt.axis('off');
```
